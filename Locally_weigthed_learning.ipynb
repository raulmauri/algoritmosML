# Locally weigthed learning

# Descripción

Prompt ChatGPT: *describe el algoritmo de ML Locally Weighted Learning*

Locally weigthed learning (LWR) es un algoritmo de aprendizaje automático que se utiliza para aproximar una función a partir de un conjunto de datos. El algoritmo funciona calculando un modelo de regresión local para cada punto de interés en el espacio de la función. El modelo de regresión local se crea utilizando solo los datos de entrenamiento que se encuentran en la vecindad del punto de interés.

La idea básica de LWR es que, en lugar de construir un modelo global para todo el espacio de la función, se puede crear un modelo local más preciso para cada punto de interés. Esto es posible porque los datos que se encuentran en la vecindad de un punto de interés tienden a ser más similares entre sí que los datos que se encuentran en otras partes del espacio de la función.

El algoritmo LWR tiene dos parámetros principales: la función de ponderación y el parámetro de suavización. La función de ponderación determina cómo se ponderan los datos de entrenamiento que se utilizan para crear el modelo de regresión local. El parámetro de suavización controla el grado de suavización del modelo de regresión local.

LWR es un algoritmo de aprendizaje no paramétrico, lo que significa que no hace ninguna suposición sobre la forma de la función que se está aproximando. Esto hace que LWR sea un algoritmo muy flexible que puede ser utilizado para aproximar una amplia gama de funciones.

LWR es un algoritmo de aprendizaje eficiente que se puede implementar de manera eficiente. Esto hace que LWR sea un algoritmo ideal para aplicaciones en las que se necesita hacer predicciones en tiempo real.

A continuación se resumen los pasos del algoritmo LWR:

1.  Seleccione un punto de interés en el espacio de la función.
2. Identifique los datos de entrenamiento que se encuentran en la vecindad del punto de interés.
3. Calcule la función de ponderación para los datos de entrenamiento.
4. Utilice la función de ponderación para crear un modelo de regresión local para el punto de interés.
5. Use el modelo de regresión local para hacer una predicción para el punto de interés.

LWR es un algoritmo de aprendizaje poderoso que puede ser utilizado para una amplia gama de aplicaciones. Es un algoritmo eficiente que se puede implementar de manera eficiente, lo que lo hace ideal para aplicaciones en las que se necesita hacer predicciones en tiempo En general, LWL es un algoritmo útil para problemas de regresión no lineales donde el conjunto de datos no es demasiado grande y el número de características es pequeño. Sin embargo, puede ser computacionalmente costoso y requiere mucha memoria para almacenar los datos de entrenamiento

# Algoritmos de entrenamiento

Prompt ChatGPT: *describe el algoritmo de entrenamoiento de Locally Weighted Learning (LWL)**



**Algoritmo de Locally Weighted Learning (LWL):**

1. **Dado:**
   - Conjunto de entrenamiento: {(x_1, y_1), (x_2, y_2), ..., (x_n, y_n)} donde x_i es un vector de características y y_i es el valor objetivo correspondiente.
   - Punto de consulta: x_q para el cual deseamos hacer una predicción.

2. **Para cada punto de consulta x_q:**
   - Calcula las distancias d(xi, x_q) entre x_q y cada punto x_i en el conjunto de entrenamiento.
   - Calcula los pesos wi usando una función de núcleo K(d), por ejemplo, K(d) = exp(-d^2).
   - Construye una matriz diagonal de pesos W donde W_ii = w_i.
   - Construye la matriz X donde cada fila es el vector de características de x_i.
   - Construye el vector y que contiene los valores objetivo correspondientes a cada x_i.

3. **Calcula el Coeficiente de Regresión βq:**
   - Calcula la matriz transpuesta de X, X^T.
   - Calcula la matriz de coeficientes βq usando la fórmula βq = (X^T W X)^-1 X^T W y.

4. **Predicción:**
   - Construye un vector x_q_extendido que es el vector de características de x_q con un 1 añadido al principio (para el término de sesgo/bias en la regresión lineal).
   - Calcula la predicción yˆq = x_q_extendido^T βq.

Este algoritmo calcula localmente un modelo de regresión lineal para el punto de consulta x_q basado en los datos de entrenamiento cercanos. Los pesos wi se utilizan para ponderar la influencia de los puntos de entrenamiento según su distancia al punto de consulta. La predicción final se realiza utilizando el modelo de regresión lineal local.

Ten en cuenta que el algoritmo presentado aquí es solo un enfoque básico. En la práctica, puede haber ajustes y optimizaciones para mejorar el rendimiento y la precisión del algoritmo.


## Supuestos y Restricciones

Prompt ChatGPT: *dame los supuestos y restricciones para  el algoritmo de  Locally Weighted Learning (LWL)*

El algoritmo de Locally Weighted Learning (LWL) se basa en ciertos supuestos y puede tener algunas restricciones que es importante considerar al utilizarlo. Aquí están los supuestos comunes y las restricciones asociadas con el algoritmo LWL:

**Supuestos:**

1. **Supuesto de Vecindario Local:** LWL asume que los puntos de datos cercanos al punto de consulta tienen una mayor influencia en la predicción. Este supuesto se refleja en la asignación de pesos basados en la distancia. Los puntos cercanos tendrán pesos más altos y los puntos lejanos tendrán pesos más bajos.

2. **Supuesto de Linearidad Local:** En su forma básica, LWL se basa en la regresión lineal local. Esto significa que se asume que dentro de una vecindad pequeña del punto de consulta, la relación entre las características y la variable objetivo es aproximadamente lineal.

3. **Supuesto de Independencia:** LWL asume que cada punto de datos es independiente y contribuye de manera única a la predicción en el punto de consulta. No se tienen en cuenta relaciones temporales ni dependencias entre puntos de datos.

**Restricciones:**

1. **Costo Computacional:** LWL puede ser costoso computacionalmente, especialmente si el conjunto de entrenamiento es grande. Calcular los pesos para cada punto y resolver el sistema de ecuaciones puede ser intensivo en términos de recursos.

2. **Elección del Núcleo y Parámetros:** La elección de la función de núcleo y sus parámetros puede tener un impacto significativo en el rendimiento del algoritmo. La elección adecuada del núcleo afecta la suavidad de las predicciones y cómo se ponderan los puntos de entrenamiento.

3. **Dimensionalidad:** En espacios de alta dimensionalidad, el rendimiento de LWL puede verse afectado debido a la maldición de la dimensionalidad. En espacios de alta dimensión, encontrar "vecinos cercanos" se vuelve menos significativo y la influencia de los puntos cercanos puede ser menos efectiva.

4. **Sesgo y Variabilidad:** Si la elección de los pesos no se gestiona adecuadamente, LWL puede sufrir de sesgo y variabilidad. Esto significa que las predicciones pueden estar demasiado influenciadas por datos cercanos y no reflejar adecuadamente patrones generales.

5. **Sobresimplificación en Modelos Lineales:** Si el supuesto de linealidad local no es válido para ciertos puntos de consulta, LWL puede proporcionar predicciones subóptimas. En tales casos, se pueden utilizar variaciones más avanzadas del algoritmo.

6. **Sensibilidad a Puntos Atípicos:** Debido a la forma en que se asignan los pesos basados en la distancia, LWL puede ser sensible a puntos atípicos o ruidosos en el conjunto de entrenamiento.

Es importante entender estos supuestos y restricciones antes de aplicar el algoritmo LWL. Dependiendo de la naturaleza de tus datos y del problema que estás abordando, es posible que necesites ajustar parámetros, considerar alternativas o evaluar otras técnicas de aprendizaje automático.

## Ejemplo en python

```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
## libreria para LWL
from sklearn.neighbors import KNeighborsRegressor

# Generar datos de ejemplo
X, y = make_regression(n_samples=100, n_features=1, noise=0.1)

# Definir el punto para el cual queremos hacer una predicción
new_point = np.array([[1.5]])

# Crear un modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo en todo el conjunto de datos
model.fit(X, y)

# Predicción usando regresión lineal estándar
prediction_standard = model.predict(new_point)

# Ahora, utilizando LWL
from sklearn.neighbors import KNeighborsRegressor

# Crear un modelo LWL con vecinos ponderados
lwl_model = KNeighborsRegressor(weights='distance')

# Entrenar el modelo LWL en todo el conjunto de datos
lwl_model.fit(X, y)

# Predicción utilizando LWL
prediction_lwl = lwl_model.predict(new_point)

print("Predicción usando regresión lineal estándar:", prediction_standard)
print("Predicción usando LWL:", prediction_lwl)
```

```python
# OTRO EJEMPLO


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('/content/tips.csv')
features = np.array(df.total_bill)
labels = np.array(df.tip)

def kernel(data, point, xmat, k):
   m,n = np.shape(xmat)
   ws = np.mat(np.eye((m)))
   for j in range(m):
      diff = point - data[j]
      ws[j,j] = np.exp(diff*diff.T/(-2.0*k**2))
   return ws

def local_weight(data, point, xmat, ymat, k):
   wei = kernel(data, point, xmat, k)
   return (data.T*(wei*data)).I*(data.T*(wei*ymat.T))

def local_weight_regression(xmat, ymat, k):
   m,n = np.shape(xmat)
   ypred = np.zeros(m)
   for i in range(m):
      ypred[i] = xmat[i]*local_weight(xmat, xmat[i],xmat,ymat,k)
   return ypred


#####################################
m = features.shape[0]
mtip = np.mat(labels)
data = np.hstack((np.ones((m, 1)), np.mat(features).T))

ypred = local_weight_regression(data, mtip, 0.5)
indices = data[:,1].argsort(0)
xsort = data[indices][:,0]

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(features, labels, color='blue')
ax.plot(xsort[:,1],ypred[indices], color = 'red', linewidth=3)
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```
