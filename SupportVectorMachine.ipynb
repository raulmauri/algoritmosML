Las Support Vector Machines (SVM) son un poderoso algoritmo de aprendizaje automático utilizado para la clasificación y la regresión. Su objetivo principal es encontrar un hiperplano en un espacio multidimensional que mejor separe las diferentes clases de datos de entrada.

Para la clasificación, se evalúa en qué lado del hiperplano de decisión caen los nuevos datos y se asigna una etiqueta de clase en función de su posición con respecto al hiperplano.

Para la regresión, se estima un valor numérico en función de la posición de los datos con respecto al hiperplano.


**Ventajas:**

Las SVM son efectivas en la clasificación de datos lineales y no lineales debido a la flexibilidad proporcionada por los kernels.
Son robustas frente al sobreajuste (overfitting) cuando se ajustan adecuadamente los parámetros.

**Desventajas:**

El entrenamiento de SVM puede ser computacionalmente costoso para grandes conjuntos de datos.
La selección de kernels y parámetros óptimos puede requerir conocimiento experto y validación cruzada.

En resumen, las Support Vector Machines son una técnica de aprendizaje automático versátil que se utiliza para resolver problemas de clasificación y regresión al encontrar un hiperplano óptimo en un espacio multidimensional que maximiza el margen entre las clases. Su capacidad para manejar datos no linealmente separables y su capacidad de generalización hacen que las SVM sean una herramienta poderosa en el campo del aprendizaje automático.

# Algoritmos de entrenamiento

Prompt ChatGPT: *describe el algoritmo de Support Vector Machine*


El algoritmo de aprendizaje que utiliza Support Vector Machines (SVM) se llama "máquina de soporte vectorial" o simplemente "SVM". El término "máquina de soporte vectorial" se refiere al modelo de clasificación o regresión construido por las SVM, que se basa en vectores de soporte y un hiperplano de separación. El proceso de entrenamiento de las SVM implica encontrar este hiperplano de manera óptima para realizar la clasificación o regresión en función de los datos de entrada y sus etiquetas.

El algoritmo de entrenamiento de Support Vector Machines (SVM) se enfoca en encontrar el hiperplano de separación óptimo que mejor discrimine entre las clases de datos. Aquí se describe el proceso de entrenamiento de las SVM:

1. **Preparación de los datos**:
   - Comienza con un conjunto de datos de entrenamiento que consiste en ejemplos etiquetados, donde cada ejemplo pertenece a una de dos clases en el caso de una clasificación binaria (SVM también se puede extender a problemas de múltiples clases).
   - Cada ejemplo de entrenamiento se representa como un vector en un espacio de características, donde cada característica corresponde a una dimensión en ese espacio.

2. **Selección del kernel**:
   - Se elige una función kernel apropiada que determine cómo se mapearán los datos de entrada al espacio de características de mayor dimensión. Los kernels comunes incluyen el kernel lineal, el kernel polinómico y el kernel radial (RBF).

3. **Definición del problema de optimización**:
   - El objetivo principal del entrenamiento de SVM es encontrar el hiperplano de separación que maximice el margen entre las clases y, al mismo tiempo, minimice la clasificación errónea. Esto se formula como un problema de optimización con restricciones.

4. **Función de coste y margen**:
   - Se define una función de coste que penaliza la clasificación errónea. La función de coste busca minimizar los errores de clasificación y maximizar el margen entre el hiperplano de separación y los ejemplos de entrenamiento más cercanos, conocidos como "vectores de soporte".
   - El parámetro de regularización C controla el equilibrio entre la maximización del margen y la minimización de errores de clasificación. Valores más altos de C permiten un margen más estrecho pero permiten un mayor número de clasificaciones erróneas, mientras que valores más bajos de C dan como resultado un margen más amplio pero penalizan más los errores de clasificación.

5. **Resolución del problema de optimización**:
   - Se utiliza un algoritmo de optimización, como el método del descenso del gradiente o el método de SMO (Sequential Minimal Optimization), para encontrar los coeficientes de los vectores de soporte y el término de sesgo del hiperplano de separación.

6. **Clasificación**:
   - Una vez que se ha encontrado el hiperplano de separación óptimo, el modelo SVM está listo para la clasificación de nuevos datos. Para un nuevo punto de datos, se evalúa en qué lado del hiperplano cae y se asigna a la clase correspondiente según su posición relativa.

En resumen, el proceso de entrenamiento de las SVM implica la optimización de un hiperplano de separación que maximiza el margen entre las clases y minimiza los errores de clasificación. Este proceso se basa en la elección de un kernel adecuado y la configuración de parámetros como C para equilibrar la precisión y la generalización del modelo. Las SVM son eficaces para separar datos en espacios de alta dimensión y son especialmente útiles cuando se necesita un clasificador sólido en problemas de clasificación lineal o no lineal.

## Supuestos y Restricciones

Prompt ChatGPT: *dame los supuestos y restricciones para  el algoritmo de  SVM*

Aquí están los supuestos y restricciones principales del algoritmo SVM:

**Supuestos:**

1. **Linealidad o no linealidad**: Los SVM pueden ser utilizados en problemas de clasificación tanto lineales como no lineales. En el caso lineal, se asume que los datos de entrada son linealmente separables en el espacio de características. Si los datos no son linealmente separables, se utiliza un kernel para mapearlos a un espacio de características de mayor dimensión donde puedan ser linealmente separables.

2. **Separabilidad**: Para SVM de clasificación, se asume que los datos de entrenamiento están etiquetados de manera que existe un hiperplano que puede separar perfectamente las clases. Esto significa que no se permite ningún error de clasificación en el conjunto de entrenamiento. En la práctica, esto puede ser un supuesto fuerte, por lo que se permite un cierto grado de error mediante la introducción del parámetro de regularización C.

**Restricciones:**

1. **Algoritmo de optimización**: SVM utiliza un algoritmo de optimización para encontrar el hiperplano de separación óptimo. Estos algoritmos pueden ser computacionalmente costosos, especialmente en conjuntos de datos grandes. Sin embargo, se han desarrollado técnicas de optimización eficientes para abordar este problema.

2. **Elección del kernel**: La elección del kernel adecuado es crucial para el rendimiento de las SVM. Seleccionar un kernel inadecuado puede llevar a un mal rendimiento del modelo. La elección del kernel y sus parámetros debe basarse en conocimiento experto o en validación cruzada.

3. **Configuración de parámetros**: SVM tiene parámetros, como el parámetro de regularización C y los parámetros específicos del kernel, que deben configurarse correctamente para obtener un buen rendimiento del modelo. Una configuración inadecuada de estos parámetros puede resultar en sobreajuste o subajuste.

4. **Problemas de escalabilidad**: En conjuntos de datos extremadamente grandes, el entrenamiento de SVM puede ser computacionalmente intensivo y requerir una cantidad significativa de memoria. En tales casos, pueden ser necesarias aproximaciones o técnicas de reducción de dimensionalidad para hacer que el algoritmo sea más escalable.

En resumen, los SVM asumen que los datos pueden ser separados por un hiperplano (o en un espacio de características de mayor dimensión) y requieren que los datos de entrenamiento estén etiquetados de manera que se pueda lograr una separación perfecta. Además, el rendimiento de SVM depende en gran medida de la elección adecuada de parámetros y la selección de un kernel adecuado. Las restricciones incluyen la necesidad de utilizar algoritmos de optimización eficientes y lidiar con problemas de escalabilidad en conjuntos de datos grandes.

## Ejemplo en python

permite clasificar si un cliente es elegible para darsele una tarjeta de credito


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Load the Universal Bank Data
df = pd.read_csv('/content/UniversalBank.csv')
df.head()
```

```python
# Dropping ID and ZIP Code columns from the dataset
df1 = df.drop(["ID","ZIP Code"], axis = 1)
df1.head()
```

```python
# Scaling the data using Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit(df1.drop('CreditCard',axis=1)).transform(df1.drop('CreditCard',axis=1))
df_scaled = pd.DataFrame(scaled, columns=df1.columns[:-1])
df_scaled.head()
```

```python
# Splitting the columns in to dependent variable (x) and independent variable (y).
x = df_scaled
y = df1['CreditCard']
```

```python
# Split data in to train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Apply SVM Model
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svc=SVC()
svc.fit(x_train, y_train)
y_pred=svc.predict(x_test)
print('Model accuracy : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'],
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='mako')
```

```python
# Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

##SVM Kernel (Tipos de Kernel)

**Linear Kernel**

```python
# Apply SVM model using Linear Kernel function
linear_classifier=SVC(kernel='linear').fit(x_train,y_train)
y_pred = linear_classifier.predict(x_test)
print('Model accuracy with linear kernel : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))
print(classification_report(y_test,y_pred))
```

**Gaussian RBF kernel**


```python
rbf_svc=SVC(kernel='rbf').fit(x_train,y_train)
y_pred = rbf_svc.predict(x_test)
print('Model accuracy with rbf kernel : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))

print(classification_report(y_test,y_pred))
```

**Polynomial Kernel**

```python
# Apply SVM model using Polynomial Kernel function
Poly_svc=SVC(kernel='poly', C=1).fit(x_train,y_train)
y_pred = Poly_svc.predict(x_test)
print('Model accuracy with rbf kernel : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))

# Classification Report
print(classification_report(y_test,y_pred))
```

**Sigmoid Kernel**

```python
#Apply SVM model using Sigmoid Kernel function
Poly_svc=SVC(kernel='sigmoid', C=1).fit(x_train,y_train)
y_pred = Poly_svc.predict(x_test)
print('Model accuracy with rbf kernel : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))

print(classification_report(y_test,y_pred))
```
