# kNN

# Descripción


K-Nearest Neighbors (KNN) es un algoritmo de aprendizaje automático no paramétrico que se utiliza para tareas de clasificación y regresión. El algoritmo funciona calculando la distancia entre un nuevo punto de datos y los puntos de datos existentes en el conjunto de entrenamiento. Los k puntos de datos más cercanos al nuevo punto de datos se denominan vecinos. El nuevo punto de datos se asigna a la clase a la que pertenecen la mayoría de sus vecinos.

El valor de k es un hiperparámetro que se debe elegir cuidadosamente. Un valor pequeño de k puede hacer que el algoritmo sea sensible al ruido, mientras que un valor grande de k puede hacer que el algoritmo sea menos preciso.

KNN es un algoritmo simple y fácil de implementar. Sin embargo, puede ser lento para entrenar y no es muy generalizable a nuevos datos.

- Las ventajas de KNN son:

Es simple y fácil de implementar.

No requiere hacer suposiciones sobre la distribución de los datos.

Es robusto al ruido.


- Las desventajas de KNN son:

Puede ser lento para entrenar.

No es muy generalizable a nuevos datos.

El rendimiento puede verse afectado por la elección del valor de k.



KNN se puede utilizar para una variedad de tareas de aprendizaje automático, como:

Clasificación de imágenes.
Clasificación de texto.
Predicción de la probabilidad de supervivencia de un paciente.
Predicción del precio de una casa.

KNN es un algoritmo intuitivo y fácil de implementar, pero puede ser computacionalmente costoso en conjuntos de datos grandes, ya que implica buscar las distancias entre todas las instancias de entrenamiento y la instancia de prueba en el momento de la predicción. Además, la elección del valor de "k" y la métrica de distancia adecuada son consideraciones importantes al usar KNN..

En resumen, el algoritmo k-NN es una técnica simple pero poderosa para la clasificación y regresión que se basa en la idea de que puntos similares tienden a tener etiquetas o valores similares.

Aplicaciones:
 * Clasificación de Documentos: k-NN puede utilizarse para clasificar documentos en categorías o temas en función de las palabras clave y características presentes en ellos.

# Algoritmo de entrenamiento

El algoritmo de aprendizaje de k-Nearest Neighbors (k-NN) es un método de clasificación y regresión que se basa en la proximidad de los puntos de datos en un espacio de características. A continuación, se presenta una descripción general del algoritmo de k-NN:

**Algoritmo de k-Nearest Neighbors (k-NN)**

1. **Elegir un valor para k:** Determine el valor de k, que representa el número de vecinos más cercanos que se utilizarán para tomar decisiones de clasificación o regresión.

2. **Elegir una métrica de distancia:** Seleccione una métrica de distancia, como la distancia euclidiana, la distancia Manhattan, etc., para calcular la proximidad entre los puntos de datos en el espacio de características.

3. **Elegir el conjunto de entrenamiento:** Use un conjunto de entrenamiento etiquetado que contenga ejemplos de datos con características conocidas y etiquetas de clase o valores objetivo.

4. **Para clasificación:**

   - Para un punto de datos de prueba, calcule la distancia entre ese punto y todos los puntos en el conjunto de entrenamiento utilizando la métrica de distancia elegida.
   - Seleccione los k puntos de entrenamiento más cercanos según la métrica de distancia.
   - Determine la clase mayoritaria entre los k vecinos más cercanos y asigne esa clase al punto de datos de prueba.

5. **Para regresión:**

   - Para un punto de datos de prueba, calcule la distancia entre ese punto y todos los puntos en el conjunto de entrenamiento utilizando la métrica de distancia elegida.
   - Seleccione los k puntos de entrenamiento más cercanos según la métrica de distancia.
   - Calcule la media o mediana de los valores objetivo de los k vecinos más cercanos y utilízala como predicción para el punto de datos de prueba.

El algoritmo de k-NN es relativamente simple y fácil de implementar. Sin embargo, es importante tener en cuenta que el rendimiento del k-NN puede depender de la elección adecuada de k, la métrica de distancia y la representación de los datos. También puede ser computacionalmente intensivo en conjuntos de datos grandes, ya que implica calcular las distancias entre el punto de prueba y todos los puntos de entrenamiento.

# Supuestos y restricciones

Supuestos y Restricciones del Algoritmo K-Nearest Neighbors (KNN):


**Supuestos:**

Proximidad en el Espacio de Características: KNN se basa en la suposición de que puntos similares en el espacio de características tienden a tener etiquetas similares. En otras palabras, si dos puntos son cercanos entre sí en términos de distancia, es más probable que pertenezcan a la misma clase.

Distribución de Datos: KNN no hace suposiciones explícitas sobre la distribución subyacente de los datos. Funciona bien en conjuntos de datos donde la frontera de decisión entre clases puede ser compleja o no lineal.


**Restricciones:**

Elección del Valor de K: La elección del valor de "K", es decir, el número de vecinos a considerar, puede afectar significativamente el rendimiento del algoritmo. Un valor muy bajo puede hacer que el modelo sea sensible al ruido, mientras que un valor muy alto puede suavizar las decisiones y hacer que se pierdan detalles importantes.

Rendimiento en Dimensiones Altas: En espacios de características de alta dimensionalidad, la intuición de distancia y proximidad puede perderse debido al fenómeno conocido como "maldición de la dimensionalidad". Esto puede llevar a resultados subóptimos o incluso ineficiencia computacional.

Sensibilidad a Datos Ruidosos y Atípicos: KNN puede ser sensible a valores atípicos y ruido en los datos. Puntos atípicos pueden influir en las decisiones si están cerca de los puntos a clasificar y si el valor de "K" es pequeño.

Eficiencia Computacional: Calcular distancias entre todos los puntos de entrenamiento y un nuevo punto puede ser costoso computacionalmente, especialmente en grandes conjuntos de datos. Se pueden aplicar técnicas como estructuras de datos de búsqueda espacial (por ejemplo, KD-Tree o Ball-Tree) para acelerar los cálculos.

Escalado de Atributos: KNN es sensible a las escalas de los atributos. Las características con rangos más amplios pueden dominar el cálculo de distancias, por lo que es importante normalizar o estandarizar los datos antes de aplicar el algoritmo.

Decisiones de Empate y Fronteras Inconsistentes: En casos donde "K" es par, puede haber empates en la votación. Esto puede llevar a fronteras de decisión inconsistentes en ciertos puntos.

En resumen, KNN es un algoritmo simple y efectivo, pero su rendimiento y precisión pueden variar según los supuestos y restricciones mencionados anteriormente. Es importante comprender estos aspectos al aplicar KNN y considerar si es adecuado para los datos y los objetivos específicos del problema.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
import seaborn as sb

%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python

2
dataframe = pd.read_csv(r"/content/reviews_sentiment.csv",sep=';')
dataframe.head(10)
```

```python
dataframe.hist()
plt.show()
```

```python
X = dataframe[['wordcount','sentimentValue']].values
y = dataframe['Star Rating'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python

n_neighbors = 7

knn = KNeighborsClassifier(n_neighbors)
knn.fit(X_train, y_train)
print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))
```

```python
pred = knn.predict(X_test)
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

```python
h = .02  # step size in the mesh

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#ffcc99', '#ffffb3','#b3ffff','#c2f0c2'])
cmap_bold = ListedColormap(['#FF0000', '#ff9933','#FFFF00','#00ffff','#00FF00'])

# we create an instance of Neighbours Classifier and fit the data.
clf = KNeighborsClassifier(n_neighbors, weights='distance')
clf.fit(X, y)

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

patch0 = mpatches.Patch(color='#FF0000', label='1')
patch1 = mpatches.Patch(color='#ff9933', label='2')
patch2 = mpatches.Patch(color='#FFFF00', label='3')
patch3 = mpatches.Patch(color='#00ffff', label='4')
patch4 = mpatches.Patch(color='#00FF00', label='5')
plt.legend(handles=[patch0, patch1, patch2, patch3,patch4])


plt.title("5-Class classification (k = 7, weights = 'DISTANCIA'")

plt.show()
```

```python
#Elegir el mejor valor de k

k_range = range(1, 20)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    scores.append(knn.score(X_test, y_test))
plt.figure()
plt.xlabel('k')
plt.ylabel('accuracy')
plt.scatter(k_range, scores)
plt.xticks([0,5,10,15,20])
```
