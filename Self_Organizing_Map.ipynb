# Self-Organizing Map

# Descripción

Prompt ChatGPT: *describe el algoritmo de ML Locally Weighted Learning*

Un Mapa Auto-organizado (Self-Organizing Map o SOM, por sus siglas en inglés), también conocido como mapa de Kohonen, es un algoritmo de aprendizaje automático no supervisado utilizado para la reducción de dimensionalidad, visualización y agrupación. Es un tipo de red neuronal artificial que mapea datos de entrada desde un espacio de alta dimensión hacia una cuadrícula de baja dimensión, al mismo tiempo que conserva las relaciones topológicas entre los puntos de datos.

En un SOM, cada neurona en la cuadrícula representa un prototipo o un vector de pesos. Durante el entrenamiento, el SOM ajusta estos vectores de pesos para mapear puntos de datos similares hacia neuronas cercanas. Las neuronas que están cerca entre sí en la cuadrícula responden de manera similar a patrones de entrada similares, lo que resulta en una representación visual de la estructura subyacente de los datos.

Los SOM se utilizan a menudo para tareas como la visualización de datos y la agrupación. Pueden ayudar a revelar patrones y relaciones dentro de datos de alta dimensión que pueden no ser evidentes de inmediato.
Es capaz de encontrar relaciones no lineales y complejas entre los datos.
SOM es útil en aplicaciones como la clasificación de imágenes, la detección de intrusiones y la minería de datos

# Algoritmos de entrenamiento

Prompt ChatGPT: *describe el algoritmo de entrenamoiento de Locally Weighted Learning (LWL)**


El algoritmo de aprendizaje de Self-Organizing Map (SOM), también conocido como Mapa Autoorganizado o Red de Kohonen, es un proceso iterativo que ajusta los pesos de las neuronas en el mapa para capturar la estructura de los datos de entrada. A continuación, se presenta una descripción general del algoritmo de entrenamiento de SOM:

1. **Inicialización:** Comienza inicializando los pesos de las neuronas en el mapa con valores aleatorios pequeños. Cada neurona en el mapa está asociada a un vector de peso de la misma dimensión que los datos de entrada.

2. **Selección de Datos:** Selecciona aleatoriamente una instancia de datos de entrada del conjunto de entrenamiento. Este será el vector de entrada que la red intentará mapear en el mapa.

3. **Competencia:** Calcula la distancia entre el vector de entrada y los vectores de peso de todas las neuronas en el mapa. Encuentra la neurona que tiene el vector de peso más cercano al vector de entrada. Esta neurona ganadora se llama "neurona ganadora" o "unidad ganadora".

4. **Actualización de Pesos:** Ajusta los pesos de todas las neuronas en el mapa utilizando una función de actualización que favorece a las neuronas más cercanas a la neurona ganadora. La actualización suele disminuir a medida que las neuronas se alejan de la neurona ganadora. Esto permite que las neuronas cercanas a la neurona ganadora se acerquen más al vector de entrada.

5. **Vecindario:** Define una vecindad alrededor de la neurona ganadora. Las neuronas dentro de esta vecindad también se actualizarán, pero su actualización será más débil a medida que se alejen de la neurona ganadora.

6. **Aprendizaje Iterativo:** Repite los pasos 2 a 5 para cada instancia de datos en el conjunto de entrenamiento durante múltiples épocas. A medida que avanza el entrenamiento, el tamaño de la vecindad suele disminuir, lo que significa que la actualización afectará a un número menor de neuronas.

7. **Resultados:** Una vez que se completa el entrenamiento, el mapa SOM se organizará de manera que regiones cercanas en el mapa representen regiones similares en los datos de entrada. Las neuronas cercanas en el mapa tendrán vectores de peso similares y, por lo tanto, representarán características similares.

El proceso iterativo de competencia y actualización gradualmente organiza las neuronas en el mapa de acuerdo con las similitudes entre los datos de entrada. A medida que se entrenan más datos, el mapa converge hacia una representación de las relaciones entre las características y los patrones en los datos.



El algoritmo se puede describir en los siguientes 4 pasos.

1. Inicialización
   Se inicializan los pesos de las neuronas en la capa del mapa.

2. Proceso competitivo
   Seleccione una muestra de entrada y busque la mejor unidad coincidente entre todas las neuronas en la cuadrícula nxm utilizando medidas de distancia.

3. Proceso cooperativo
   Encuentre las neuronas de proximidad de BMU por función de vecindad.

4. Proceso de adaptación
   Actualice las ponderaciones de la BMU y de los vecinos desplazando los valores hacia el patrón de entrada.
   Si se alcanza el recuento máximo de iteraciones de entrenamiento, salga. De lo contrario, incremente el recuento de iteraciones en 1 y repita el proceso desde 2.


## Supuestos y Restricciones

Prompt ChatGPT: *dame los supuestos y restricciones para  el algoritmo de  Locally Weighted Learning (LWL)*

El algoritmo "Self-Organizing Map" (SOM), también conocido como Mapa Autoorganizado o Red de Kohonen, tiene ciertos supuestos y restricciones que es importante tener en cuenta al utilizarlo. Aquí están algunos de los supuestos y restricciones comunes asociados con el algoritmo SOM:

**Supuestos:**

1. **Topología Ordenada:** Se asume que los datos en el espacio de entrada tienen una estructura topológica ordenada. Esto significa que los patrones similares en el espacio de entrada deberían estar cerca entre sí en el mapa SOM resultante.

2. **Preservación de la Topología:** El algoritmo se basa en la presunción de que la topología de los datos originales se conservará en el mapa SOM. Es decir, las regiones vecinas en el espacio de entrada se mapearán a regiones vecinas en el mapa.

3. **Exploración de Características:** Se supone que el mapa SOM será capaz de descubrir características y relaciones entre los datos en función de su estructura topológica. Esto permite la identificación de agrupamientos y patrones en los datos.

**Restricciones:**

1. **Alta Dimensionalidad de los Datos:** El rendimiento del algoritmo SOM puede degradarse en espacios de alta dimensionalidad. La maldición de la dimensionalidad puede afectar la representación de los datos en el mapa SOM.

2. **Elección de Parámetros:** La elección adecuada de parámetros, como el tamaño del mapa, la tasa de aprendizaje y la cantidad de épocas de entrenamiento, es crucial para el rendimiento del algoritmo. Una elección inadecuada puede llevar a un mapa poco informativo.

3. **Sensibilidad a Inicialización:** El resultado final del mapa SOM puede variar según la inicialización de los pesos de las neuronas. Diferentes inicializaciones pueden conducir a diferentes organizaciones del mapa.

4. **Requerimientos Computacionales:** Dependiendo del tamaño del conjunto de datos y del mapa SOM, el algoritmo puede ser intensivo en recursos computacionales, especialmente en términos de memoria y tiempo de cómputo.

5. **No Linealidades Complejas:** Aunque SOM puede capturar relaciones y agrupamientos no lineales en los datos, su capacidad para modelar relaciones extremadamente complejas puede ser limitada.

6. **Interpretación Subjetiva:** Interpretar el mapa SOM y sus características puede ser subjetivo. Las decisiones sobre la cantidad de neuronas, la disposición y la interpretación de las regiones en el mapa dependen del conocimiento del usuario.

7. **No Supervisión:** SOM es un algoritmo de aprendizaje no supervisado, lo que significa que no se basa en valores objetivo para entrenar. Esto puede limitar su capacidad para realizar predicciones precisas o para modelar relaciones específicas.

Es importante tener en cuenta estos supuestos y restricciones al utilizar el algoritmo SOM en la práctica. Aunque es una técnica poderosa para visualización y organización de datos, su rendimiento puede variar según el conjunto de datos y la configuración de parámetros.

## Ejemplo en python

Este ejemplo utiliza el conjunto de datos de flores Iris para crear un mapa SOM y visualizar cómo las características de las flores se organizan en el mapa. Las neuronas ganadoras se marcan con colores según las clases reales de las flores. El mapa SOM muestra cómo las características se agrupan y organizan en el espacio bidimensional del mapa.

```python
!pip install MiniSom
import numpy as np
from minisom import MiniSom
import matplotlib.pyplot as plt
from sklearn import datasets

# Cargar el conjunto de datos de flores
iris = datasets.load_iris()
data = iris.data
labels = iris.target

# Normalizar los datos
data = (data - np.mean(data, axis=0)) / np.std(data, axis=0)

# Configuración del mapa SOM
map_size = (10, 10)
input_len = data.shape[1]
learning_rate = 0.5
sigma = 2.0

# Crear y entrenar el mapa SOM
som = MiniSom(map_size[0], map_size[1], input_len, sigma=sigma, learning_rate=learning_rate)
som.random_weights_init(data)
som.train_random(data, 1000)  # Entrenar con 1000 épocas

# Etiquetas para las clases
class_names = ['Setosa', 'Versicolor', 'Virginica']

# Asignar cada instancia a su neurona ganadora en el mapa
winning_neurons = [som.winner(x) for x in data]

# Visualizar el mapa SOM
plt.figure(figsize=(10, 10))
plt.pcolor(som.distance_map().T, cmap='bone_r')  # Mostrar la distancia entre neuronas
plt.colorbar()

# Marcar las neuronas ganadoras con colores según las clases reales
for i, (x, label) in enumerate(zip(winning_neurons, labels)):
    plt.text(x[0] + 0.5, x[1] + 0.5, str(class_names[label]),
             color=plt.cm.rainbow(label / 2.5), fontdict={'weight': 'bold', 'size': 9})

plt.xticks(np.arange(map_size[0] + 1))
plt.yticks(np.arange(map_size[1] + 1))
plt.show()
```

En este ejemplo, el SOM se entrena con el conjunto de datos de autenticación de billetes disponible en el sitio web del repositorio UCI ML . El archivo de datos contiene 1372 filas, cada una con 4 características y 1 etiqueta. Las 4 características son valores numéricos sin nulo. La etiqueta es un valor entero binario.


Este es  un ejemplo de clasificacion

```python
import numpy as np
from numpy.ma.core import ceil
from scipy.spatial import distance #distance calculation
from sklearn.preprocessing import MinMaxScaler #normalisation
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score #scoring
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from matplotlib import animation, colors



data_file = "data_banknote_authentication.txt"
data_x = np.loadtxt(data_file, delimiter=",", skiprows=0, usecols=range(0,4) ,dtype=np.float64)
data_y = np.loadtxt(data_file, delimiter=",", skiprows=0, usecols=(4,),dtype=np.int64)


# train and test split
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.2, random_state=42)
print(train_x.shape, train_y.shape, test_x.shape, test_y.shape) # check the shapes
```

```python
# Helper functions

# Data Normalisation
def minmax_scaler(data):
  scaler = MinMaxScaler()
  scaled = scaler.fit_transform(data)
  return scaled

# Euclidean distance
def e_distance(x,y):
  return distance.euclidean(x,y)

# Manhattan distance
def m_distance(x,y):
  return distance.cityblock(x,y)

# Best Matching Unit search
def winning_neuron(data, t, som, num_rows, num_cols):
  winner = [0,0]
  shortest_distance = np.sqrt(data.shape[1]) # initialise with max distance
  input_data = data[t]
  for row in range(num_rows):
    for col in range(num_cols):
      distance = e_distance(som[row][col], data[t])
      if distance < shortest_distance:
        shortest_distance = distance
        winner = [row,col]
  return winner

# Learning rate and neighbourhood range calculation
def decay(step, max_steps,max_learning_rate,max_m_dsitance):
  coefficient = 1.0 - (np.float64(step)/max_steps)
  learning_rate = coefficient*max_learning_rate
  neighbourhood_range = ceil(coefficient * max_m_dsitance)
  return learning_rate, neighbourhood_range
```

```python
# hyperparameters
num_rows = 10
num_cols = 10
max_m_dsitance = 4
max_learning_rate = 0.5
max_steps = int(7.5*10e3)

# num_nurons = 5*np.sqrt(train_x.shape[0])
# grid_size = ceil(np.sqrt(num_nurons))
# print(grid_size)
```

```python
#main function

train_x_norm = minmax_scaler(train_x) # normalisation

# initialising self-organising map
num_dims = train_x_norm.shape[1] # numnber of dimensions in the input data
np.random.seed(40)
som = np.random.random_sample(size=(num_rows, num_cols, num_dims)) # map construction

# start training iterations
for step in range(max_steps):
  if (step+1) % 1000 == 0:
    print("Iteration: ", step+1) # print out the current iteration for every 1k
  learning_rate, neighbourhood_range = decay(step, max_steps,max_learning_rate,max_m_dsitance)

  t = np.random.randint(0,high=train_x_norm.shape[0]) # random index of traing data
  winner = winning_neuron(train_x_norm, t, som, num_rows, num_cols)
  for row in range(num_rows):
    for col in range(num_cols):
      if m_distance([row,col],winner) <= neighbourhood_range:
        som[row][col] += learning_rate*(train_x_norm[t]-som[row][col]) #update neighbour's weight

print("SOM training completed")
```

```python
# collecting labels

label_data = train_y
map = np.empty(shape=(num_rows, num_cols), dtype=object)

for row in range(num_rows):
  for col in range(num_cols):
    map[row][col] = [] # empty list to store the label

for t in range(train_x_norm.shape[0]):
  if (t+1) % 1000 == 0:
    print("sample data: ", t+1)
  winner = winning_neuron(train_x_norm, t, som, num_rows, num_cols)
  map[winner[0]][winner[1]].append(label_data[t]) # label of winning neuron
```

```python
# construct label map
label_map = np.zeros(shape=(num_rows, num_cols),dtype=np.int64)
for row in range(num_rows):
  for col in range(num_cols):
    label_list = map[row][col]
    if len(label_list)==0:
      label = 2
    else:
      label = max(label_list, key=label_list.count)
    label_map[row][col] = label

title = ('Iteration ' + str(max_steps))
cmap = colors.ListedColormap(['tab:green', 'tab:red', 'tab:orange'])
plt.imshow(label_map, cmap=cmap)
plt.colorbar()
plt.title(title)
plt.show()
```

```python
# test data

# using the trained som, search the winning node of corresponding to the test data
# get the label of the winning node

data = minmax_scaler(test_x) # normalization

winner_labels = []

for t in range(data.shape[0]):
 winner = winning_neuron(data, t, som, num_rows, num_cols)
 row = winner[0]
 col = winner[1]
 predicted = label_map[row][col]
 winner_labels.append(predicted)

print("Accuracy: ",accuracy_score(test_y, np.array(winner_labels)))
```

Tenemos algunos desafíos al aplicar el mapa autoorganizado a un escenario del mundo real. Primero, si no tenemos un conjunto de datos etiquetado, no podemos medir la pérdida. No tenemos medios para validar qué tan confiable es el mapa entrenado. La calidad del mapa depende en gran medida de las características de los propios datos. El preprocesamiento de datos mediante normalización es esencial para el algoritmo basado en la distancia. Un análisis previo del conjunto de datos también es importante para comprender la distribución de los puntos de datos. Especialmente para los datos de alta dimensionalidad donde la visualización no es posible, podemos utilizar otras técnicas de reducción de dimensionalidad como PCA y descomposición de valores singulares (SVD).

Además, es posible que el entrenamiento no tenga éxito si la forma del mapa topológico no es relevante para la distribución de los puntos de datos en el espacio latente. Aunque utilizamos una cuadrícula en nuestro ejemplo, debemos diseñar la formación del mapa con cuidado. Uno de los enfoques recomendables es utilizar la proporción de las varianzas explicadas de los dos primeros componentes principales del PCA. Sin embargo, si el tiempo lo permite, vale la pena probar los diferentes hiperparámetros para torneado fino.

En términos del costo computacional del algoritmo, la complejidad del tiempo de entrenamiento depende del número de iteraciones, el número de características y el número de neuronas. Originalmente, el SOM está diseñado para el aprendizaje secuencial, pero hay casos en los que se prefiere el método de aprendizaje por lotes. A medida que aumenta el tamaño de los datos para big data, puede ser necesario investigar algoritmos de aprendizaje más eficaces. En nuestro ejemplo, utilizamos la función de vecindad rectangular, llamada burbuja para simplificar. Para las iteraciones de entrenamiento, podríamos monitorear la formación de mapas autoorganizados y examinar cómo se aprende el mapa durante el ciclo. La reducción del número de pasos de entrenamiento impacta directamente en el número de cálculos.
